```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:/Users/syeda_ojd0o2c/Documents/Project 2/crime data.csv')
df.head()
```

# Data Cleaning

```python
# Check for missing values
print(df.isnull().sum())
```

# Basic Statistics

```python
# Summary statistics for each region
regions = ['Punjab', 'Sindh', 'KP', 'Balochistan', 'Islamabad', 'Railways', 'G.B', 'AJK']
for region in regions:
    print(f"\nStatistics for {region}:")
    print(df[region].describe())
```

```python
# Total crime in Pakistan
df['Total Crime'] = df[regions].sum(axis=1)
print("Total Crime Statistics:")
print(df['Total Crime'].describe())
```

```python
# Trends over time
df.groupby('Year')['Total Crime'].sum().plot(kind='line')
plt.title('Total Crime in Pakistan Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Crime')
plt.show()
```

```python
# Regional Analysis
df.groupby('Year')[regions].sum().plot(kind='line')
plt.title('Crime in Different Regions Over the Years')
plt.xlabel('Year')
plt.ylabel('Crime')
plt.legend(loc='upper right')
plt.show()
```

```python
# Specific Offence Analysis
offences = df['Offence'].unique()
for offence in offences:
    offence_data = df[df['Offence'] == offence]
    offence_data.groupby('Year')['Total Crime'].sum().plot(kind='line')
    plt.title(f'Total Crime for {offence} Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Total Crime')
    plt.show()
```
